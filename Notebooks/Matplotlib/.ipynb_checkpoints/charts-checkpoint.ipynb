```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sn
```

```python
values = [0, 1, 2, 3, 4, 5, 6, 7]
squares = [v**2 for v in values]

plt.plot(values, squares)
plt.show()
```

```python
values = list(range(30))
pows = [pow(v, 3) for v in values]
plt.scatter(values, pows)
plt.show()
```

```python
plt.scatter(values, pows, c=pows, cmap=plt.cm.Blues, edgecolor='none', s=10)
plt.show()
```

```python
values = list(range(10))
squares = [pow(v, 2) for v in values]
cubes = [pow(v, 3) for v in values]

fig, (ax1, ax2) = plt.subplots(1, 2, sharey=True)
ax1.scatter(values, squares)
ax1.set_title("Squares")
ax2.plot(values, cubes)
ax2.set_title("Cubes")
plt.show()
```

```python
data = {
    "Year": [2001, 2002, 2003, 2004, 2005, 2006],
    "Sales": [100, 150, 130, 180, 210, 230]
}
df = pd.DataFrame(data)
df.plot(x="Year", y="Sales", kind="line")
plt.title("Sales")
plt.show()
```

```python
df.plot(x="Year", y="Sales", kind="bar")
plt.show()
```

```python
data = {
    "Year": [2001, 2002, 2003, 2004, 2005, 2006],
    "Sales": [100, 150, 130, 180, 210, 230]
}
df = pd.DataFrame(data)
df.plot(x="Year", y="Sales", kind="area")
plt.title("Sales")
plt.show()
```

```python
data = {
    "Year": [2001, 2002, 2003, 2004, 2005, 2006],
    "Sales": [100, 150, 130, 180, 210, 230]
}
df = pd.DataFrame(data)
df.plot(y="Sales", kind="pie", labels=df["Year"], autopct="%1.1f%%")
plt.title("Sales")
plt.show()
```

```python
data = {
    "Age": [25, 28, 32, 41, 42, 42, 50, 59, 60]
}
df = pd.DataFrame(data)
df.plot(y="Age", kind="hist")
plt.title("Age")
plt.show()
```

```python
data = {
    "Category": ["Cat1", "Cat1", "Cat2", "Cat2", "Cat3", "Cat3", "Cat2"],
    "Value": [10, 20, 25, 45, 45, 50, 50]
}
df = pd.DataFrame(data)
df.boxplot(by="Category", column="Value")
plt.title("Value distribution by category")
plt.show()
```

```python
data = {
    "Month": ["Jan", "Feb", "Mar", "Apr", "May"],
    "Week 1": [6, 10, 12, 14, 17],
    "Week 2": [8, 12, 13, 14, 18],
    "Week 3": [9, 11, 14, 16, 20],
    "Week 4": [10, 15, 10, 17, 22]
}
df = pd.DataFrame(data)
df.set_index("Month", inplace=True)
df
```

```python
sn.heatmap(df, cmap="coolwarm", annot=True)
plt.title("Weekly Temperatues")
plt.show()
```
