```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("jersey-rainfall-1984-to-2021.csv", index_col=0)
df.tail()
```

```python
df.plot(y="YearTotalmm", kind="area")
plt.show()
```

```python
df.plot(y=["Apr", "May"], kind="area")
plt.show()
```

```python
# Transposition. Columns go instead rows and vice versa
df.T.tail()
```

```python
dft = df.drop(columns="YearTotalmm").T
dft.plot(y=[1901, 1902, 2001, 2002], kind="bar")
plt.show()
```
