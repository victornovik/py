## 5. Calculate Total and Average Quantities

```python
# Column - products (4 products), Row - orders (2 orders)
qty = np.array([[5,3,2,7], 
                [10,6,3,9]])
qty
```

```python
qty_per_product = np.sum(qty, axis=0)
qty_per_product
```

```python
avg_qty_per_order = np.mean(qty, axis=1)
avg_qty_per_order
```

## 6. Calculate Average and Maximum Rating

```python
# Column - category (4 categories), Row - product (3 products)
# Each product is rated by 4 categories
rating = np.array([
    [4.5, 3.2, 2.5, 5.0], 
    [4.3, 3.8, 1.0, 4.8],
    [2.0, 3.6, 4.7, 0.5]
])
rating
```

```python
avg_rating_per_category = np.round(np.mean(rating, axis=0), 1)
avg_rating_per_category
```

```python
max_rating_per_category = np.round(np.max(rating, axis=0), 1)
max_rating_per_category
```

## 8. Generate Sample Stock Data (3D array)

```python
companies = ["Google", "Microsoft", "Apple"]
days = ["Mon 1 April", "Tue 2 April"]
types = ["Open", "Close", "High", "Low"]

# 3x2x4
prices = np.round(np.random.random((len(companies), len(days), len(types))), 3)
prices
```

```python
for i, company in enumerate(companies):
    print(f"{company} prices:")
    for j, day in enumerate(days):
        print(f"\tDay {day}:")
        for k, t in enumerate(types):
            print(f"\t\t{t} Price: {prices[i,j,k]}")    
```
