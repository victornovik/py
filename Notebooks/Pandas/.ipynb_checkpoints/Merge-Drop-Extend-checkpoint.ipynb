```python
import pandas as pd
import numpy as np

users = {
    "ID": [1,2,3,4,5],
    "Username": ["Victor", "Alex", "Tanya", "Lev", "Timofey"],
    "Age": [18, 35, 25, 38, 45],
    "Date": pd.to_datetime(["2032-01-01", "2032-02-15","2032-04-25","2032-06-21","2032-09-15"]),
    "Posts": [150, 230, 80, 420, None],
    "Reputation": [500, 720, 200, 940, 500]
}
df = pd.DataFrame(users)
df
```

## Extend

```python
df["Language"] = "English"
df
```

```python
df["Active"] = [True, False, False, True, True]
df
```

## Merge

```python
comments = {
    "Comments": [70, 30, 45, 55, 80]
}
df2 = pd.DataFrame(comments)
df2

# Assign custom row indices
# df2 = pd.DataFrame(comments, index=["a", "b", "c", "d", "e"])
```

```python
df = pd.concat([df, df2], axis=1)
df
```

## Drop

```python
# Returns new dataframe, does not change the original dataframe
df.drop(["Language"], axis=1)
```

```python
# Change the original dataframe
if ("Language" in df.columns):
    df.drop(["Language"], axis=1, inplace=True)
df
```

```python
# Remove rows
df.drop([2, 3])
```
