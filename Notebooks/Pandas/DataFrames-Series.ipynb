```python
users = {
    "ID": [1,2,3,4,5],
    "Username": ["Victor", "Alex", "Tanya", "Lev", "Timofey"],
    "Age": [18, 35, 25, 38, 45],
    "Date": pd.to_datetime(["2032-01-01", "2032-02-15","2032-04-25","2032-06-21","2032-09-15"]),
    "Posts": [150, 230, 80, 420, None],
    "Reputation": [500, 720, 200, 940, 500]
}
df = pd.DataFrame(users)
df
```

```python
# D = EX² - E²X
age = df["Age"]
squared_avg = age.mean()**2
avg_of_squares = age.apply(lambda x: x**2).sum() / len(age)
var = avg_of_squares - squared_avg
std = var ** 0.5
std
```

```python
# D = E(X - EX)²
age = df["Age"]
avg = age.mean()
var = age.apply(lambda x: (x - avg)**2).sum() / len(age)
std = var ** 0.5
std
```
